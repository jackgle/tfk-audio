#### Prepare spectrograms
---

```python
specgen = soundkit.dataprep.spec.SpecGenerator(sample_rate=16000,
                                               stft_window_seconds=0.05,
                                               stft_hop_seconds=0.01,
                                               db_limits=(None, None))
```

```python
# process a single file
wav, _ = soundkit.dataprep.audio.load('tmp/positive/42545_1/3308066_1.9-4.19.wav')
spec = specgen.wav_to_spec(wav)
```

```python
# process all .wav files in a folder
specgen.process_folder('./tmp/', limit=None)
```

```python
# check examples
specgen.plot_examples(path='./tmp/positive/4931_1/')
```
